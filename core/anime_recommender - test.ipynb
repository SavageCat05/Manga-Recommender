```python
class GenreOneHotEncoder(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.unique_genres = None

    def fit(self, X, y=None):
        # Extract unique genres from the 'Genres' column
        all_genres = X['Genres'].str.split(', ').explode().unique()
        self.unique_genres = sorted(all_genres)  # Sort for consistent column order
        return self

    def transform(self, X):
        # Create one-hot encoded columns with prefix "Genre_"
        genre_columns = {f"Genre_{genre}": [] for genre in self.unique_genres}
        
        for genres in X['Genres']:
            genre_set = set(genres.split(', '))
            for genre in self.unique_genres:
                genre_columns[f"Genre_{genre}"].append(1 if genre in genre_set else 0)
        
        # Convert to DataFrame
        genre_df = pd.DataFrame(genre_columns)
        
        # Concatenate original dataframe with one-hot encoded columns
        X_encoded = pd.concat([X.reset_index(drop=True), genre_df], axis=1)
        return X_encoded

# Example usage
data = {
    "anime_id": [1, 2, 3],
    "Name": ["Cowboy Bebop", "Trigun", "Monster"],
    "Genres": ["Action, Sci-Fi", "Action, Adventure, Sci-Fi", "Drama, Mystery, Suspense"]
}

df = pd.DataFrame(data)

# Initialize and apply the transformer
encoder = GenreOneHotEncoder()
encoder.fit(df_anime)
df_encoded = encoder.transform(df_anime)




```

```python
import asyncio
import aiohttp
import pandas as pd
import nest_asyncio

# Apply nest_asyncio for environments like Jupyter Notebook
nest_asyncio.apply()

# Function to fetch anime data asynchronously with exponential backoff
async def fetch_anime_data_with_backoff(session, anime_id, retries=5):
    """
    Fetch anime data asynchronously with exponential backoff in case of 'Too Many Requests' error.
    """
    url = f"https://api.jikan.moe/v4/anime/{anime_id}"
    delay = 4  # Initial delay in seconds for backoff

    for attempt in range(retries):
        try:
            async with session.get(url) as response:
                if response.status == 429:  # Too many requests error
                    print(f"Rate limit hit for anime_id {anime_id}. Retrying in {delay} seconds...")
                    await asyncio.sleep(delay)
                    delay *= 2  # Exponential backoff
                    continue
                
                if response.status == 404:  # Not Found error
                    print(f"Anime ID {anime_id} not found (404). Skipping...")
                    return anime_id, "Unknown", "Unknown"
                
                response.raise_for_status()
                data = await response.json()
                title = data["data"].get("title", "Unknown")
                year = data["data"].get("year", "Unknown")
                return anime_id, title, year
        except Exception as e:
            print(f"Error fetching data for anime_id {anime_id}: {e}")
            return anime_id, "Unknown", "Unknown"
    
    # If all retries fail
    return anime_id, "Unknown", "Unknown"

# Function to fill missing Release_Year values asynchronously with rate limiting and backoff
async def fill_and_clean_anime_years_async(df):
    """
    Fill missing Release_Year values asynchronously with backoff handling and rate limiting.
    """
    not_found_dict = {}
    missing_rows = df[df["Release_Year"].isna()].copy()
    
    # Semaphore to limit the number of concurrent requests (set to 1 or 2 for stricter rate limits)
    semaphore = asyncio.Semaphore(1)

    async def fetch_with_semaphore(anime_id):
        async with semaphore:
            async with aiohttp.ClientSession() as session:
                return await fetch_anime_data_with_backoff(session, anime_id)

    # Create tasks for all missing rows
    tasks = [fetch_with_semaphore(row["anime_id"]) for _, row in missing_rows.iterrows()]
    
    # Gather results concurrently with global delay between requests
    results = []
    for task in asyncio.as_completed(tasks):
        results.append(await task)
        await asyncio.sleep(0.5)  # Add global delay between requests
    
    # Process results
    for anime_id, title, year in results:
        if year not in ["Unknown", "Not found", None]:
            df.loc[df["anime_id"] == anime_id, "Release_Year"] = int(year)
        else:
            not_found_dict[anime_id] = title
    
    # Drop rows where year couldn't be found
    df = df[~df["anime_id"].isin(not_found_dict.keys())].reset_index(drop=True)
    
    return df, not_found_dict

# Wrapper function to run the async function directly on your DataFrame
def update_dataframe_with_release_years(df):
    """
    Wrapper to run the asynchronous function on a given DataFrame.
    """
    return asyncio.run(fill_and_clean_anime_years_async(df))



# Example usage of the code
# if __name__ == "__main__":
#     # Example DataFrame with anime IDs and missing Release_Year values
#     data = {
#         "anime_id": [6553, 9045, 20471, 33736, 35500, 37437],  # Replace with your actual anime IDs
#         "Release_Year": [None, None, None, None, None, None]   # Missing years
#     }
    
#     df_anime = pd.DataFrame(data)
    
#     # Run the function to fill missing Release_Year values
#     df_anime_cleaned, unfound_animes = run_async_fill_and_clean(df_anime)
    
#     # Save removed anime info (optional)
#     pd.DataFrame(list(unfound_animes.items()), columns=["anime_id", "title"]).to_csv("unfound_anime.csv", index=False)

#     # Print results
#     print("Cleaned DataFrame:")
#     print(df_anime_cleaned)

#     print("\nAnimes with unresolved Release_Year issues:")
#     print(unfound_animes)





```

```python
import pandas as pd

def create_training_data(merged_df, df_user, df_anime):
    """
    Creates the training dataset by merging user, anime, and interaction data.

    Args:
        merged_df (pd.DataFrame): DataFrame containing user-anime interactions with `user_id`, `anime_id`, and `rating`.
        df_user (pd.DataFrame): DataFrame containing user features with `user_id`.
        df_anime (pd.DataFrame): DataFrame containing anime features with `anime_id`.

    Returns:
        XU (pd.DataFrame): User feature matrix.
        XM (pd.DataFrame): Anime feature matrix.
        target_variable (pd.Series): Target variable (ratings).
    """
    # Step 1: Map user_id and anime_id to integers
    user_mapping = {user: idx for idx, user in enumerate(df_user['user_id'].unique())}
    anime_mapping = {anime: idx for idx, anime in enumerate(df_anime['anime_id'].unique())}

    # Apply mappings to merged_df
    merged_df['user_id'] = merged_df['user_id'].map(user_mapping)
    merged_df['anime_id'] = merged_df['anime_id'].map(anime_mapping)

    # Apply mappings to df_user and df_anime
    df_user['user_id'] = df_user['user_id'].map(user_mapping)
    df_anime['anime_id'] = df_anime['anime_id'].map(anime_mapping)

    # Step 2: Merge df_user and df_anime with merged_df
    merged_with_user = pd.merge(merged_df, df_user, on='user_id', how='left')
    merged_with_anime = pd.merge(merged_with_user, df_anime, on='anime_id', how='left')

    # Step 3: Extract XU (user features), XM (anime features), and target variable (rating)
    XU = merged_with_anime[df_user.columns]  # User features
    XM = merged_with_anime[df_anime.columns]  # Anime features
    target_variable = merged_with_anime['rating']  # Target variable

    return XU, XM, target_variable

```
