```python
# Display information about the DataFrame
df_user.info()

# Filter users with complete information
complete_users = df_user.dropna()

# Display the filtered DataFrame
print("Users with complete information:")

complete_users.head(10)
```

```python
complete_users.columns
columns_list = complete_users.columns.tolist()
print(columns_list)
main_users = complete_users[['Mal ID', 'Username', 'Gender', 'Birthday', 'Location','Mean Score','Completed']]
```

```python
# main_users["Location"].unique()  not all users have location data, so it is not relevant for the recommender system
```

```python
# Extract the year directly using string slicing
main_users['Birth_Year'] = main_users['Birthday'].str[:4]

# Display the updated DataFrame
print(main_users[['Birthday', 'Birth_Year']])
```

```python
main_users.info()
```

```python
# drop the birthday column as it is not relevant for the recommender system
main_users = main_users.drop(columns=['Birthday'])
main_users.info()
main_users.isna().sum()
```

```python
main_users = main_users.drop(columns=['Location'])
# location, was anyways useless. so removed :>
main_users.isna().sum()
```

## Necessary user demographics have been added

```python
df_score.head(10)
```

```python
df_anime.head(10)
```

```python
# Filter rows where user_id matches Mal ID and Username matches
filtered_df = df_score[
    (df_score['user_id'].isin(main_users['Mal ID'])) & 
    (df_score['Username'].isin(main_users['Username']))
]

# Drop columns with all zero values
filtered_df = filtered_df.loc[:, (filtered_df != 0).any(axis=0)]

# Display the filtered DataFrame
print(filtered_df)
```
